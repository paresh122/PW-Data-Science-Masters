Q6. List down some best practices for exception handling in python.

### Some of the Best practice for exception is :
- Use always specific exception 
    - like if we know which exception our code raise use only that handler
- Use always valid message in our exception
- Always try to log all the error 
    - make us better to understand our code behaviour
- avoid using multiple exception statment
    - if we know what kind of error our code block could raise
- made proper documentation
